```python
import pandas as pd
```

```python
sentence_label = pd.read_csv('./local_data/sentence-label.csv', index_col=0)
sentence_label
```

```python
pneumonia_text = ' .'.join(sentence_label[sentence_label.Pneumonia == 1].Reports.values.tolist())
len(pneumonia_text)
```

```python
import nltk
from nltk.collocations import *


finder = TrigramCollocationFinder.from_words(pneumonia_text.split(' '))

finder.apply_freq_filter(10)

pneumonia_trigram = {k: v for k, v in finder.ngram_fd.items() if 'pneumonia' in k}
len(pneumonia_trigram)
```

```python
dict(sorted(pneumonia_trigram.items(), key=lambda item: item[1], reverse=True))
```

```python
dict(sorted(finder.ngram_fd.items(), key=lambda item: item[1], reverse=True))
```
